```python
import pandas as pd
import matplotlib.pyplot as plt

returns_data = pd.read_csv("../Data/Returns.csv", parse_dates = [0], index_col = 0)

returns = returns_data.iloc[:,0]

cum_returns = (1 + returns).cumprod()

annualized_returns = returns.mean() * 252
annualized_volatility = returns.std() * (252 ** 0.5)
sharpe = annualized_returns/annualized_volatility if annualized_volatility !=0 else float("nan")
max_drawdown = (cum_returns/cum_returns.cummax()- 1).min()

metrics = pd.DataFrame({
    "Annualized_Return": [annualized_returns],
    "Annualized_Volatility": [annualized_volatility],
    "Sharpe_Ratio": [sharpe],
    "Max_DrawDown": [max_drawdown]
})
```

```python
metrics
```

```python
plt.figure(figsize = (14,6))
plt.title("Cumulative Returns")
plt.xlabel("Time")
plt.ylabel("Growth of $1")
plt.plot(cum_returns)
plt.grid(False)
```
